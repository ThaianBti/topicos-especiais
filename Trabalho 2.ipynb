```python
from bokeh.plotting import figure

from bokeh.io import output_notebook, show

import pandas as pd

from bokeh.plotting import ColumnDataSource

from bokeh.layouts import gridplot

from bokeh.models.widgets import Panel

# Import Tabs from bokeh.models.widgets
from bokeh.models.widgets import Tabs

data = pd.read_csv("CO2_passenger_cars_v12_editado.csv", encoding = 'latin2', low_memory=False)

# Create ColumnDataSource: source
AUDI = ColumnDataSource(data[data["Mk"] == "AUDI"])
CHRYSLER = ColumnDataSource(data[data["Mk"] == "CHRYSLER"])
SEAT = ColumnDataSource(data[data["Mk"] == "SEAT"])

#figures
p1 = figure(x_axis_label='m (kg)', y_axis_label='e (g/km)')
p2 = figure(x_axis_label='m (kg)', y_axis_label='e (g/km)')
p3 = figure(x_axis_label='m (kg)', y_axis_label='e (g/km)')


#circles
p1.circle('m (kg)','e (g/km)', source=AUDI, color='red', legend="AUDI")
p1.circle('m (kg)','e (g/km)', source=CHRYSLER, color='blue', legend="CHRYSLER")

p2.circle('m (kg)','e (g/km)', source=AUDI, color='red', legend="AUDI")
p2.circle('m (kg)','e (g/km)', source=SEAT, color='blue', legend="SEAT")

p3.circle('m (kg)','e (g/km)', source=CHRYSLER, color='red', legend="CHRYSLER")
p3.circle('m (kg)','e (g/km)', source=SEAT, color='blue', legend="SEAT")

# Assign the legend to the bottom right: p.legend.location
p1.legend.location = 'bottom_right'
p2.legend.location = 'bottom_right'
p3.legend.location = 'bottom_right'

# Fill the legend background with the color 'lightgray': p.legend.background_fill_color
p1.legend.background_fill_color='lightgray'
p2.legend.background_fill_color='lightgray'
p3.legend.background_fill_color='lightgray'



# Create tab1 from plot p1: tab1
tab1 = Panel(child=p1, title='AUDI x CHRYSLER')

# Create tab2 from plot p2: tab2
tab2 = Panel(child=p2, title='AUDI x SEAT')

# Create tab2 from plot p3: tab2
tab3 = Panel(child=p3, title='CHRYSLER X SEAT')


# Create a Tabs layout: layout
layout = Tabs(tabs=[tab1,tab2,tab3])

# Call the output_notebook() 
output_notebook()

# Display the plot
show(layout)
```
